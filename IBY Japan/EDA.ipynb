## Loading dataframes

## Performing Basic EDA
---

```python

# Function to perform basic EDA on a dataframe
def basic_eda(df, name):
    print(f"\n\n--- {name} Dataset ---\n")
    
    # Shape of the dataset
    print("Shape:", df.shape)
    
    # First few rows
    print("\nFirst 5 rows:")
    display(df.head())
    
    # Data types and missing values
    print("\nData Info:")
    display(df.info())
    
    print("\nMissing Values:")
    print(df.isnull().sum())
    
    # Basic statistical summary
    print("\nStatistical Summary:")
    display(df.describe())
    
   

# Perform EDA on each dataset
basic_eda(emotion_df, 'Emotion')
basic_eda(gaze_df, 'Gaze')
basic_eda(transcript, 'Transcript')

```

## Stacked Bar Chart for Sentiment Analysis (Positive, Neutral, Negative) vs ID
___

```python
# Group the necessary data for plotting
stacked_data = transcript[['id', 'positive', 'neutral', 'negative']]

# Set 'id' as the index for easier plotting
stacked_data.set_index('id', inplace=True)

# Plotting the stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'blue', 'red'])

# Adding labels and title
plt.title('Sentiment Analysis: Positive, Neutral, and Negative vs ID')
plt.xlabel('ID')
plt.ylabel('Sentiment Score')
plt.legend(title="Sentiment")
plt.tight_layout()

# Show plot
plt.show()
```

## Stacked Bar Chart for Confidence vs Hesitance (Confident and Hesitant Scores) vs ID
___

```python
# Group the necessary data for plotting
stacked_data = transcript[['id', 'confident', 'hesitant']]

# Set 'id' as the index for easier plotting
stacked_data.set_index('id', inplace=True)

# Plotting the stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6), color=['orange', 'purple'])

# Adding labels and title
plt.title('Confidence vs Hesitance: Confident and Hesitant Scores vs ID')
plt.xlabel('ID')
plt.ylabel('Score')
plt.legend(title="Sentiment")
plt.tight_layout()

# Show plot
plt.show()
```
